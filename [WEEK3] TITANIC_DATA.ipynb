```
data["몸무게"].mean()
```

```
data["지점"] = data["지점"].replace("강남구", "강남")
data
```

```
data["몸무게"] = data["몸무게"].fillna(75.8)
data
```

```
data["몸무게(prev)"] = data["몸무게"].shift(+1) #어제
data["몸무게(next)"] = data["몸무게"].shift(-1) #내일

data
```

## 컬럼 추가 & 수정

```
data["지역"] = "부산"

data
```

```
data["결제"] = [True, True, True, True, True, True, True, True, True]

data
```

```
data["몸무게(gram)"] = data["몸무게"] * 1000

data
```

```
data["과체중"] = data["몸무게"] > 75.8

data
```

```
data.loc[data["몸무게"] > 75.8, "상태"] = "과체중"
data.loc[data["몸무게"] <= 75.8, "상태"] = "정상"

data
```
